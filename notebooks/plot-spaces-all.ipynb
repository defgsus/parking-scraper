```python
def plot_vbar_time(x, y, x_label="time", y_label="y", width_minutes=15, plot_width=930, plot_height=400):
    p = figure(plot_width=plot_width, plot_height=plot_height, y_axis_label=y_label, x_axis_type="datetime")
    p.xaxis.formatter = get_datetime_formatter()
    p.vbar(x, width_minutes*60*1000, y, alpha=.5)
    show(p)
    
def plot_line_time(x, y, true_y, x_label="time", y_label="y", line_width=2, plot_width=930, plot_height=400):
    p = figure(plot_width=plot_width, plot_height=plot_height, y_axis_label=y_label, x_axis_type="datetime")
    p.xaxis.formatter = get_datetime_formatter()
    
    max_y = 0.
    line_x, line_y = [], []
    for px, py in zip(x, y):
        if py is None:
            if line_x:
                p.line(line_x, line_y, line_width=line_width)
            line_x = []
            line_y = []
        else:
            line_x.append(px)
            line_y.append(py)
            max_y = max(max_y, py)
    
    if line_x:
        p.line(line_x, line_y, line_width=line_width)
        
    p.y_range = Range1d(-.1*max_y, 1.1*max_y)
    
    show(p)
```

```python
def plot_sources(sources, plot_height=200, min_std=0.):
    storage = Storage()
    places = storage.load_sources_arrays(sources)
    for place in places:
        true_y = [y for y in place["y"] if y is not None]
        if true_y:
            std = np.std(true_y, dtype=np.float32)
            if std > min_std:
                print("-"*23, place["place_id"], "-"*23)
                plot_line_time(place["x"], place["y"], true_y, plot_height=plot_height)
```

```python
for source in DataSources().sources:
    print("\n", source["source_id"], "\n")
    plot_sources(DataSources().filtered(source["source_id"]), min_std=10)
```
